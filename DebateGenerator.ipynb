```python
import spacy
from transformers import pipeline
from IPython.display import display, Markdown
```

```python
nlp = spacy.load("en_core_web_sm")
generator = pipeline("text-generation", model="gpt2")
```

```python
def generate_text(prompt, max_len=150):
    result = generator(prompt, max_length=max_len, num_return_sequences=1)[0]["generated_text"]
    return result.strip()

def extract_points(text, max_points=3):
    doc = nlp(text)
    return [sent.text for sent in doc.sents][:max_points]

def display_md(text):
    display(Markdown(text))

def generate_detailed_debate(topic):
    display_md(f"# 🗣️ Debate on: **{topic}**")

    # Introduction
    intro_prompt = f"Write an engaging introduction to the debate topic: '{topic}'"
    intro = generate_text(intro_prompt, 120)
    display_md(f"### 🔰 Introduction\n{intro}")

    # For the topic
    for_prompt = f"Write a detailed argument supporting the topic: '{topic}'. Include 3 strong supporting points."
    for_argument = generate_text(for_prompt, 200)
    for_points = extract_points(for_argument, 3)

    display_md("### ✅ Arguments FOR the Topic")
    display_md(for_argument)
    for idx, pt in enumerate(for_points, 1):
        display_md(f"- **Point {idx}:** {pt}")

    # Against the topic
    against_prompt = f"Write a detailed argument opposing the topic: '{topic}'. Include 3 strong opposing points."
    against_argument = generate_text(against_prompt, 200)
    against_points = extract_points(against_argument, 3)

    display_md("### ❌ Arguments AGAINST the Topic")
    display_md(against_argument)
    for idx, pt in enumerate(against_points, 1):
        display_md(f"- **Point {idx}:** {pt}")

    # Conclusion
    conclusion_prompt = f"Write a balanced conclusion for the debate on: '{topic}'"
    conclusion = generate_text(conclusion_prompt, 100)
    display_md("### 🧾 Conclusion")
    display_md(conclusion)
```

```python
topic = input("Enter a debate topic: ")
generate_detailed_debate(topic)
```
